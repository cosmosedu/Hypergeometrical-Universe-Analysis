# q parameter with LCDM model
Try to identify universe is accelerating or Decelerating  with dark energy Dominating universe

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.linspace(0.1,2,1000)
# we know q(a)=0.5*omega_m - omega_L
omega_m0 = 0.3
omega_l0 = 0.7

omega_m = omega_m0/(omega_m0 + omega_l0 )*a**3
omega_l = (omega_l0)*a**3/(omega_m0 + omega_l0)*a**3

q= 0.5*omega_m - omega_l

plt.figure()
plt.plot(a, q, label="q(a)")
plt.axhline(0, linestyle='--')
plt.xlabel("Scale factor a")
plt.ylabel("Deceleration parameter q")
plt.title("Day 7: Deceleration → Acceleration Transition (ΛCDM)")
plt.legend()
plt.grid(True)
plt.show()


```

# Results/learning
The negative value of the q-parameter indicates that the universe is in
an accelerated expansion phase at late times.
