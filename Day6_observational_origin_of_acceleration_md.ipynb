# Day 6: Observational Origin of Cosmic Acceleration

In the previous days, I focused on diagnosing cosmic expansion using
quantities such as the scale factor, acceleration, and the deceleration
parameter q. While models like ΛCDM predict a transition to accelerated
expansion at late times, it is essential to establish whether this
behavior is actually observed in the real universe.

The primary observational evidence for late-time cosmic acceleration
comes from Type Ia supernova measurements. Type Ia supernovae are treated
as standard candles because their intrinsic luminosities are nearly
uniform. This allows their distances to be inferred from their observed
brightness.

When the inferred distances of distant supernovae are compared with
their redshifts, the supernovae are found to be farther away than
expected in a purely decelerating universe. This indicates that the
expansion rate of the universe has increased at late times.

Therefore, cosmic acceleration is not a theoretical assumption but an
observationally inferred result. Cosmological models such as ΛCDM aim to
provide a consistent explanation for this observed behavior.
