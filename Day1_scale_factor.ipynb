```python
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace (0.1,10,200)
a = t**(2/3)
#why only 2/3 . directly from physics +Einstein's equations and 2/3 is a standard mathematical form that comes from from specific physical assumption which I accept for now and I will unpack gradually.

plt.figure()
plt.plot(t,a)
plt.xlabel ("cosmic time (t)")
plt.ylabel ("scale factor a(t)")
plt.title("Matter-Dominate universe Expansion")
plt.grid(True)
plt.show
```
