# Day 6: Deceleration Parameter (q)

**Goal:**  
To determine whether the universe is accelerating or decelerating
using the q-parameter, a clean and dimensionless diagnostic.

```python
import numpy as np
import matplotlib.pyplot as plt

t= np.linspace(0.1,13.8,1500)
a= t**(2/3)
a_dot = (2/3) * t**(-1/3)
a_ddot = -(2/9)*t**(-4/3)
q = -(a*a_ddot)/(a_dot**2)
print("q parameter of universe:",q)

plt.figure()
plt.plot(t, q)
plt.axhline(0, linestyle='--')   # reference line
plt.xlabel("Cosmic Time (t)")
plt.ylabel("Deceleration Parameter q")
plt.title("q-parameter for Matter Dominated Universe")
plt.grid(True)
plt.show()
```

# Results/ learning

q parameter is dimensionless quantity which indicates universe acceleration when this quantity is positive then universe is decelerating and when it is negative that means universe is accelerating.  it is independent of scale and numerical noise
