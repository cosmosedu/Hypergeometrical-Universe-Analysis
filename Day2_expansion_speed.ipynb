# Expansion speed of universe
basic question  arises in mind why not velocity and I get my answere In cosmology, velocities are coordinate-dependent and can be misleading due to the expansion of space itself. Therefore, instead of defining velocities of galaxies, we study the evolution of the scale factor and its normalized rate of change, the Hubble parameter, which provides a coordinate-independent description of cosmic expansion.
here small question in mind gives me big direction in my research.

```python
import numpy as np
import matplotlib.pyplot as plt

t= np.linspace(0.1,13.8,150)
#0 can lead singularity
#13.8 roughly age of universe
#150 for better slicing
a = t**2/3
a_dot = np.gradient(a,t)
print(a_dot)

H= a_dot/a


plt.plot(a, H)
plt.xlabel("scale factor")
plt.ylabel("H(t)")
plt.title("H(t) vs scale factor : graph 1")
plt.show()

plt.plot(t, H)
plt.xlabel("cosmic Time")
plt.ylabel("H(t)")
plt.title("H(t) vs cosmic time : graph 2 ")
plt.show()
```

# Results
In the early universe, the expansion rate was very high as shown by large values of H(t).
As cosmic time and scale factor increase, H(t) decreases.
This means the universe continues to expand, but the expansion becomes slower with time.
Hence, the Big Bang represents rapid expansion of space, not an explosion.
