# Day 3: Acceleration vs Deceleration of the Universe

**Aim:**  
To understand the difference between expansion rate and acceleration using the scale factor.

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0.1,13.8,150)
a = t**2/3

#double derivative of scale factor
a_ddot = -(2/9)*t**(-4/3)
print(a_ddot)

plt.plot(t,a_ddot)
plt.xlabel("time")
plt.ylabel("acceleration")
plt.title("expansion acceleration")
plt.show()
```
